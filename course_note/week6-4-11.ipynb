```python
stdata = pd.read_csv(r'data\students.csv')
print(stdata[-3:])
```

```python
stdata.plot(kind='scatter', x='Height', y='Weight', title='Student Body Shape', marker='*', grid=True, xlim=[150, 200],
            ylim=[40, 80], label='(Height,Weight)')
```

```python
data_male = stdata[stdata['Gender'] == 'male']
data_female = stdata[stdata['Gender'] == 'female']

plt.figure()
plt.scatter(data_male['Height'], data_male['Weight'], color='red', marker='s', label='Male')
plt.scatter(data_female['Height'], data_female['Weight'], color='blue', marker='^', label='Female')

plt.xlim(150, 200)
plt.ylim(40, 80)
plt.title('Student Body')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.grid()
plt.legend(loc='upper right')
```

```python
data_population = pd.read_csv(r'data\population.csv', index_col='Year')
print(data_population)
data_boy_girl = data_population[['Boys', 'Girls']]
mean = np.mean(data_boy_girl, axis=0)
std = np.std(data_boy_girl, axis=0)
print("年度出生数均值：", mean)
print("年度出生数标准差：", std)
```

```python
fig = plt.figure()
plt.subplots_adjust(wspace=0.6)
ax1 = fig.add_subplot(1, 2, 1)
mean.plot(kind='bar', yerr=std, color='cadetblue', title='Average of Births', rot=45, ax=ax1)
ax2 = fig.add_subplot(1, 2, 2)
mean.plot(kind='barh', xerr=std, color='cadetblue', title='Average of Births', ax=ax2)
```

```python
data_boy_girl.plot(kind='bar', title='Births of Boys & Girls')
data_boy_girl.plot(kind='bar', stacked=True, title='Births of Boys & Girls')
```

```python
ad_data = pd.read_csv(r'data/advertising.csv')
print("列名:", ad_data.columns.tolist())  # 输出: ['Unnamed: 0', 'TV', 'Weibo', 'WeChat', 'Sales']
selected_ad_data = ad_data[['TV', 'Weibo', 'WeChat']]
datasum = selected_ad_data.sum()
print("\n各渠道总支出:\n", datasum)
plt.figure(figsize=(6, 6))
ax = datasum.plot(
    kind='pie',
    title='Advertising Expenditure',
    fontsize=14,
    explode=[0, 0.1, 0],  # 3个值对应3个渠道（突出Weibo 10%）
    shadow=True,
    startangle=60,
    autopct='%1.1f%%',
    colors=['#1f77b4', '#ff7f0e', '#2ca02c']  # 添加颜色更美观
)
plt.ylabel('')
plt.tight_layout()
plt.show()
```

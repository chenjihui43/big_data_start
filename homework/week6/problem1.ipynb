```python
#1. 2012~2017年我国人均可支配收入为[1.47, 1.62, 1.78, 1.94, 2.38, 2.60](单位：万元)。按照要求绘制以下图形。
```

```python
#（1）模仿例4-1和4-3，绘制人均可支配收入折线图。用小矩形标记数据点，红色虚线，用注解标注最高点，图标题“Income chart”，设置坐标轴标题，最后将图形保存为JPG文件。一维数组访问。
```

```python
import matplotlib.pyplot as plt
# 设置中文字体和解决负号显示问题
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
plt.rcParams['axes.unicode_minus'] = False    # 解决负号显示为方块的问题
plt.figure()
yearData=['2012','2013','2014','2015','2016','2017']
incomeData=[1.47, 1.62, 1.78, 1.94, 2.38, 2.60]
plt.plot(yearData,incomeData,color="red",linestyle="--",marker="s")
plt.title('Income chart')
plt.xlabel('Year')
plt.ylabel('Income(RMB Ten Thousands)')
plt.grid(True)

# 找到最高点
max_income = max(incomeData)
max_index = incomeData.index(max_income)
max_year = yearData[max_index]

# 添加注解
plt.annotate(
    f'Highest:{max_income}',
    xy=(max_index, max_income),          # 注解位置（x, y）
    xytext=(10, 10),                     # 文字偏移量（单位为像素）
    textcoords='offset points',          # 偏移方式
    arrowprops=dict(arrowstyle='->',    # 箭头样式
                    connectionstyle='arc3,rad=0.2',  # 弧线连接
                    color='blue'),       # 箭头颜色
    fontsize=10,
    bbox=dict(boxstyle='round,pad=0.3', fc='white', ec='gray')  # 背景框
)

plt.show()
```

```python
# （2）模仿例4-2，使用多个子图分别绘制人均可支配收入的折线图、箱须图以及柱状图。
```

```python
import matplotlib.pyplot as plt
yearData=['2012','2013','2014','2015','2016','2017']
incomeData=[1.47, 1.62, 1.78, 1.94, 2.38, 2.60]
fig=plt.figure()
area1=fig.add_subplot(2,2,1)
area2=fig.add_subplot(2,2,2)
area3=fig.add_subplot(2,1,2)

area1.plot(yearData,incomeData)
area1.set_title('line chart')
area1.set_xlabel('Year')
area1.set_ylabel('Income')


area2.boxplot(incomeData)
area2.set_title('box-whisker chart')
area2.set_xlabel('2012~2017')
area2.set_ylabel('Income')
area2.set_xticks([])  # 隐藏 x 轴刻度

# area3 绘制柱状图
area3.bar(yearData, incomeData, color='skyblue', edgecolor='black')
area3.set_title('bar chart')
area3.set_xlabel('Year')
area3.set_ylabel('Income')
# 设置纵坐标刻度范围和间隔
yticks = [0, 0.5, 1.0, 1.5, 2.0, 2.5,3.0]
area3.set_yticks(yticks)
area3.set_ylim(0, 3.0)

# 可选：调整布局，防止标签重叠
plt.tight_layout()
plt.show()
```
